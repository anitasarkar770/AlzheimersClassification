```python
# Blur Images

cn_path = "gdrive/MyDrive/UNCC2024/Project/ADNI2/CN/jpgs/"
cn_path_blur = "gdrive/MyDrive/UNCC2024/Project/ADNI2/CN/blurjpgs/"

cn_images = Path(cn_path).glob ("*.jpg")

for file in cn_images:
  img = cv2.imread (file)
  blur_img = cv2.GaussianBlur(img, (5,5), 0)
  filename = os.path.basename(file)
  newpath = os.path.join(cn_path_blur,filename)
  cv2.imwrite(newpath, blur_img)

new_images = Path(cn_path_blur).glob ("*.jpg")
print (len(list(new_images)))
```

```python
mci_path = "gdrive/MyDrive/UNCC2024/Project/ADNI2/MCI/jpgs/"
mci_path_blur = "gdrive/MyDrive/UNCC2024/Project/ADNI2/MCI/blurjpgs/"


mci_images = Path(mci_path).glob ("*.jpg")

for file in mci_images:
  img = cv2.imread (file)
  blur_img = cv2.GaussianBlur(img, (5,5), 0)
  filename = os.path.basename(file)
  newpath = os.path.join(mci_path_blur,filename)
  cv2.imwrite(newpath, blur_img)

new_images = Path(mci_path_blur).glob ("*.jpg")
print (len(list(new_images)))
```

```python
ad_images =glob.glob(ad_dstn + "/**/*", recursive=True)
print (len(ad_images))
for file in ad_images:
  if file.endswith(".dcm"):
    ds = dicom.dcmread(file)
    img = Image.fromarray(ds.pixel_array)
    if img.mode != 'RGB':
      img = img.convert('RGB')
    img.save(ad_path + os.path.basename(file) + ".jpg")

```
